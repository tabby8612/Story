# Lesson 100 : Topic-100. Iterative Structures: Loop Execution (Example-II)


1. In this example, we see how the loop for factorial works.



```python
def factorial(num):
    factor = 1
    iternation = 1
    while (iternation <= num):
        factor = factor * iternation
        iternation = iternation + 1
    print(factor)

factorial(2)
```

# Lesson 106 : Topic-106. Algorithm efficiency

1. In computer science when we talk about algorithm efficiency then we mean by the time it took to search element

2. Suppose we have a data of 30k students then we can use sequential search (searching every element from beginning) or binary search (searching element from middle and reducing list)

3. If the registar uses sequential search then suppose if one element search takes 10 milliseconds then it will take 30,000 * 10/1000 = 300 seconds or 5 mins that can be embrassing

4. However, in binary search algorithm, every time the function run it divides the list in half. So, 15 * 10 / 1000 = 0.15 seconds

5. Both the above examples of sequential and binary search are examples of worst case as we are considering if the element found at the end of list.

6. Now suppose if the element found at the middle of the list then still sequential search takes 15000 * 10/1000 = 150 seconds or 2.5 mins which is still more than worst sensior of binary search
